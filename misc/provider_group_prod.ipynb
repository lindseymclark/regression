# Functions

```python
#create some useful functions and variables
#string matching function with fuzzywuzzy. Can also use Levenshtein Distance Ratio.
#I ended up using the SequenceMatcher in python, https://docs.python.org/2/library/difflib.html, which is similar.

def match_fuzz_string(str2, str1):
    global ratio
    ratio = fuzz.ratio(str1.lower(),str2.lower())
    return ratio

#define a minimum ratio for matching
min_fuzz_ratio = 75

#define a function that creates a list of Provider_Entity_ids and the correct number of repeats
provider_id = []
repeats = []
def getlists(provider_list):
    for item in provider_list:
        global provider_id
        global repeats
        a = provider_list.index(item)
        provider_id.append(a)
        b = len(item)
        repeats.append(b)
    return provider_id, repeats 

#get lists of lists of provider ids. Expands the two lists from the function above to get a list of list of all
#the provider entity id's so I can append to dataframe.
list_of_provider_entity_ids= []
def get_provider_id_list(ids, vals):
    global list_of_provider_entity_ids
    for i,j in zip(ids, vals):
        a1 = [i]
        a2 = a1*j
        list_of_provider_entity_ids.append(a2)
    return list_of_provider_entity_ids   
```

# Create the groups

```python
#get a list of the unique Billing__Provider_name
provider_list = provider_data.Billing_Provider_Name.unique()
provider_list = provider_list.tolist()
```

```python
#loop to get the groups in a list of lists
start_time = datetime.now() 
result = []
for provider in provider_list:
    if(len(result)==0):
        result.append([provider])
    else:
        for i in range(0,len(result)):
            score=match_fuzz_string(provider,result[i][0])
            if(score<90):
                if(i==len(result)-1):
                    result.append([provider])    
            else:
                if(score>=90):
                    result[i].append(provider)
time_elapsed = datetime.now() - start_time 
print('Time elapsed (hh:mm:ss.ms) {}'.format(time_elapsed))
```
